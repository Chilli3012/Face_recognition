```python
import cv2
import numpy as np
```

```python
#initiaze the camera
cap=cv2.VideoCapture(0)
```

```python
#for face detection
face_cascade=cv2.CascadeClassifier("../cdMLAI/HaarCascade/haarcascade_frontalface_alt.xml")
```

```python
##the running loop

skip=0
face_data=[]#to store the faces.to save memory save faces in grayscale
dataset_path='/Users/anuragsingh/Desktop/cdMLAI/Data/Face/'#path were faces are stored

file_name=input("Enter the name of the person : ")


while True:
    ret,frame=cap.read()#to read what is captured on the cam
    
    
    if ret==False:##like in any case if the camera does not open try it again
        continue
        
        
    faces=face_cascade.detectMultiScale(frame,1.3,5)
    print(faces)
    
    #now if there are multiple faces then we need to sort the face in the order of there area.cause we only want to store the largest face
    faces=sorted(faces,key=lambda f:f[2]*f[3],reverse=True)
    
    
    for x,y,w,h in faces:#we iterate in reverse because the last face will be the largest one
        #x,y,w,h=face
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        
        #Extract (Crop out the required face):Region of Interest
        offset=10#to give the extra region of a particular size in the image
        face_section=frame[y-offset:y+h+offset,x-offset:x+w+offset]
        face_section=cv2.resize(face_section,(100,100))
        
        skip+=1
        if(skip%10==0):
            face_data.append(face_section)
            print(len(face_data))
            
            
            
        
    cv2.imshow("Frame",frame)
    cv2.imshow("Face_Section",face_section)
        
    
    
    
    ##part to exit the loop
    key_pressed=cv2.waitKey(1) & 0xFF
    if key_pressed==ord('q'):
        break
        
        
  

 ## now we convert aur face list array into a numpy array
face_data=np.asarray(face_data)
face_data=face_data.reshape((face_data.shape[0],-1))
print(face_data.shape)



## now we save this data into system file
np.save(dataset_path+file_name+'.npy',face_data)
print("data successfully saved"+dataset_path+file_name+'.npy')


cap.release()
cv2.destroyAllWindows()
        
```
